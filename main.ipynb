```python
%matplotlib inline

import logging
from Trainer import Trainer
from Models.PIAE import PIAE
import utils

dataset = 'mnist'
# dataset = 'cifar10'

utils.set_logging(logging.INFO) 
kwargs = utils.read_config(dataset)

# initialize the model
model = PIAE(lambda_=20, dist='MI', **kwargs)
# put the model in the Trainer
trainer = Trainer(model, **kwargs)
# begin loading the existing model and start the training process
trainer.load().train()
# calculate the subspace score
trainer.subspace_score()
```
